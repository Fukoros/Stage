```python
import pandas as pd
import numpy as np
import sklearn.tree as tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import os.path
from os import path
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

# 1Â° Load the data

```python
data_raw = pd.read_csv("Data/miniloan-decisions-100K.csv")
```

```python
data_raw = data_raw.drop(columns=["name"])
```

```python
data_raw["approval"].iloc[0]
```

```python
false_data = data_raw[data_raw["approval"] == " false"]
true_data = data_raw[data_raw["approval"] == " true"]
```

```python
data_equilibrium = true_data.append(false_data[:len(true_data)])
```

# 2 Select the thresholds

```python
%%time

def returnSplitValue(stringToSplit):
    res = []
    tp = ""
    spliters = [("lt", "< "), ("lte", "<= "), ("gt", "> "), ("gte", ">= ")]
    i=0
    
    while i < len(spliters):
        splited = stringToSplit.split(spliters[i][1]) 
        if len(splited) == 2:
            return  spliters[i][0], float(splited[1])
        i += 1
    print("C'est casse")
    return None 


res = []
for key in data_equilibrium.keys():
    if key != "approval":
        print("\t\t"+key)
        X = data_equilibrium[[key]]

        Y = data_equilibrium["approval"]

        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
        
        model = tree.DecisionTreeClassifier(max_depth=1)
        model.fit(X_train, y_train)    

        print(f"Train Score : {accuracy_score(y_train, model.predict(X_train))}")
        print(f"Test Score : {accuracy_score(y_test, model.predict(X_test))}")
        
        if accuracy_score(y_train, model.predict(X_train)) > 0.65 and accuracy_score(y_test, model.predict(X_test)) > 0.65:
            res.append((key,)+ returnSplitValue(tree.export_text(model, spacing=1, feature_names=[key]).split("\n")[0]))
print(res)
```

```python
%%time

data_raw_wth_name = data_raw.drop(columns=["name", "approval"])

Y = data_raw["approval"]

X_train, X_test, y_train, y_test = train_test_split(data_raw_wth_name, Y, test_size=0.33, random_state=42)

#You should not run it as it takes too much time.
model = tree.ExtraTreeClassifier()
model.fit(X_train, y_train)

#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=data_raw_wth_name.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()

print(f"Train Score : {accuracy_score(y_train, model.predict(X_train))}")
print(f"Test Score : {accuracy_score(y_test, model.predict(X_test))}")
```

```python
def findBestFeatures(X, y, toFind, model):
    
    class_n_best_features_real = toFind if toFind <= len(X.columns) else len(X.columns) 
    
    bestFeatures = []
    
    while class_n_best_features_real > 0 :
         
        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
        
        model.fit(X_train, y_train)

        #plot graph of feature importances for better visualization
        feat_importances = pd.Series(model.feature_importances_, index=X.columns)
        bestFeatures.append(feat_importances.sort_values(ascending=False).index[0])
        
        X = X.drop(columns=bestFeatures[-1])
        class_n_best_features_real -= 1
    
    return bestFeatures
```

```python
data_raw_wth_name = data_raw.drop(columns=["name", "approval"])

Y = data_raw["approval"]

findBestFeatures(data_raw_wth_name, Y, 10, tree.ExtraTreeClassifier())
```
