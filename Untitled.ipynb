```python
from sklearn.datasets import load_iris
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("./test/train.csv")
```

```python
X = data[data.columns[:-1]].dropna(axis=1)
Y = data[data.columns[-1]]
```

```python
le = LabelEncoder()
le.fit(X)
```

```python
for column in X.dtypes.index:
    if X.dtypes[column] == "object":
        X[column] = LabelEncoder().fit(X[column]).transform(X[column])
```

```python
clf = tree.DecisionTreeClassifier(random_state=0)
iris = load_iris()
```

```python
fig, _axs = plt.subplots(nrows=1, ncols=1, figsize=(20,10))

clf = clf.fit(iris.data, iris.target)
tree.plot_tree(clf)  
```

```python
fig, _axs = plt.subplots(nrows=1, ncols=1, figsize=(20,10))

clf = clf.fit(X[X.columns[:10]], Y)
tree.plot_tree(clf)  
a=2
```
